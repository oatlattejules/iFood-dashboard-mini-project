```python
# import packages
import pandas as pd
import numpy as np
```

```python
# read data
data = pd.read_csv("ifood_df.csv")
data.isnull().sum()
```

```python
# total spending per customer 
mnt_cols = [col for col in data.columns if col.startswith("Mnt")]
data["TotalSpending"] = data[mnt_cols].sum(axis=1)
```

```python
# total campaign acceptance 
cmp_cols = [col for col in data.columns if col.startswith("AcceptedCmp")]
data["TotalCampaignAcceptance"] = data[cmp_cols].sum(axis=1)
```

```python
# add Customer Years var
data["Customer_Years"] = data["Customer_Days"] / 365
```
